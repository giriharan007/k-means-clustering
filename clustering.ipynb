```python
data = ((data  - data.min())/(data.max()-data.min())) * 9+1 #min-max scaling!(1.subtract min value from value(sets min val = 0)2.divide value by the range3.multiply with range.4.add one)
data.describe()
```

```python
def random_centroids(data,k):

    centroids =[]
    for i in range(k):
        
        centroid = data.apply(lambda x :float(x.sample())) # initilizing random centroids
        centroids.append(centroid)
    return pd.concat(centroids, axis = 1)
```

```python
centroids = random_centroids(data , 5)
```

```python

centroids
```

```python
def get_labels(data, centroids):
    distances = centroids.apply(lambda x:np.sqrt(((data - x)**2).sum(axis = 1)))
    return distances.idxmin(axis =1)
```

```python
labels = get_labels(data, centroids)
```

```python
labels.value_counts()
```

```python
#calculating geometric mean(center of the cluster)
def new_centroids(data, labels, k):
    return data.groupby(labels).apply(lambda x:np.exp(np.log(x).mean())).T
```

```python
from sklearn.decomposition  import PCA
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
def plot_clusters(data, labels, centroids, iteration):
    pca = PCA(n_components=2)   #transforming data into 2d
    data_2d = pca.fit_transform(data) 
    centroids_2d = pca.transform(centroids.T)
    clear_output(wait= True)
    plt.title(f'Iteration{iteration}')
    plt.scatter(x=data_2d[:,0], y = data_2d[:,1], c = labels)
    plt.scatter(x=centroids_2d[:,0], y = centroids_2d[:,1])
    plt.show()

```

```python
max_iterations = 100
k= 3 #no of clusters

centroids = random_centroids(data, k)
old_centroids = pd.DataFrame()

iteration = 1

while iteration < max_iterations and not centroids.equals(old_centroids):
    old_centroids = centroids

    labels = get_labels(data, centroids)
    centroids = new_centroids(data, labels, k)
    plot_clusters(data, labels, centroids, iteration)
    iteration +=1

```
